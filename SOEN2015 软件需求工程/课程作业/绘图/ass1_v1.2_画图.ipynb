```python
import numpy as np
from pyecharts import OverLap
from pyecharts.charts import *
from pyecharts.components import Table
from pyecharts import options as opts
from pyecharts.commons.utils import JsCode
import random
import datetime
from pyecharts.globals import CurrentConfig

CurrentConfig.ONLINE_HOST = "https://cdn.kesci.com/lib/pyecharts_assets/"
```

```python
theme_list = ["chalk",
              "dark",
              "essos",
              "infographic",
              "light",
              "macarons",
              "purple-passion",
              "roma",
              "romantic",
              "shine",
              "vintage",
              "walden",
              "westeros",
              "white",
              "wonderland"]
```

```python
x_data_1 = [2015, 2016, 2017, 2018, 2019, 2020]
x_data_2 = ["2016", "2017", "2018", "2019", "2020"]
y_data_1 = [885, 1725, 6020, 11800, 21875, 36750]
y_data_2 = [round((y_data_1[i+1] / y_data_1[i] - 1) * 10000) / 100 for i in range(len(y_data_1) - 1)]
print(y_data_1)
print(y_data_2)
```

```python
bar = (
    Bar(init_opts=opts.InitOpts(theme="infographic"))
    .add_xaxis(x_data_1)
    .add_yaxis("分布式存储需求量: 套", y_data_1, z=0)
    .extend_axis(
        yaxis=opts.AxisOpts(
            axislabel_opts=opts.LabelOpts(formatter="{value}%")
        )
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(title="2015-2020年中国分布式存储行业需求总量及增速"),
        legend_opts=opts.LegendOpts(type_="scroll", pos_left="60%", orient="horizontal"),
        xaxis_opts=opts.AxisOpts(
            type_="category",
            splitline_opts=opts.SplitLineOpts(
                is_show=True, linestyle_opts=opts.LineStyleOpts(opacity=0.3)
            ),
            splitarea_opts=opts.SplitAreaOpts(is_show=True,
                                                areastyle_opts=opts.AreaStyleOpts(opacity=1))
        ),
        yaxis_opts=opts.AxisOpts(
            type_="value",
            splitline_opts=opts.SplitLineOpts(
                is_show=True, linestyle_opts=opts.LineStyleOpts(opacity=0.3)
            )
        )
    )
)

line = (Line()
        .add_xaxis(x_data_2)
        .add_yaxis("同比增长: %", y_data_2, yaxis_index=1)
)

bar.overlap(line)
bar.set_colors(["#FC97AF", "#65A8C8"])
bar.render("a.html")
bar.render_notebook()

```
